## 3. What is a Monte Carlo simulation (also known as "Monte Carlo method")

Speak to enough scientists, and you hear the words “Monte Carlo” a lot. “We ran the Monte Carlos,” a researcher will say. What does that mean?

The scientists are referring to Monte Carlo simulations, a statistical technique used to model probabilistic (or “stochastic”) systems and establish the odds for a variety of outcomes. The concept was first popularized right after World War II, to study nuclear fission mathematician Stanislaw Ulam coined the term in reference to an uncle who loved playing the odds at the Monte Carlo casino (then a world symbol of gambling, like Las Vegas today).

To get a handle on a Monte Carlo simulation, first consider a scenario where we do not need one: to predict events in a simple, linear system. If you know the precise direction and velocity at which a shot put leaves an Olympic athlete’s hand, you can use a linear equation to accurately forecast how far it will fly. This case is a deterministic one, in which identical initial conditions will always lead to the same outcome.

The world, however, is full of more complicated systems than a shot-put toss. In these cases, the complex interaction of many variables — or the inherently probabilistic nature of certain phenomena — rules out a definitive prediction. So a Monte Carlo simulation uses essentially random inputs (within realistic limits) to model the system and produce probable outcomes

## 4. Monte Carlo simulation to a real-life system

Today there are multiple types of Monte Carlo simulations, used in fields from particle physics to engineering, finance and more.

<u>Here are some uses of MCs:</u>

<b>Sampling</b>. Here the objective is to gather information about a random object by observing many realizations of it. An example is simulation modeling, where a random process mimics the behavior of some real-life system, such as a production line or telecommunications network. Another example is found in Bayesian statistics, where Markov chain Monte Carlo (MCMC) is often used to sample from a posterior distribution.


<b>Estimation</b>. In this case the emphasis is on estimating certain numerical quantities related to a simulation model. An example in the natural setting of Monte Carlo techniques is the estimation of the expected throughput in a production line. An example in the artificial context is the evaluation of multi-dimensional integrals via Monte Carlo techniques by writing the integral as the expectation of a random variable.


<b>Optimization</b>. The MCM is a powerful tool for the optimization of complicated objective functions. In many applications these functions are deterministic and randomness is introduced artificially in order to more efficiently search the domain of the objective function. Monte Carlo techniques are also used to optimize noisy functions, where the function itself is random — for example, the result of a Monte Carlo simulation.

## 5.Calculation of $\pi$ Using the Monte Carlo Method

We assume a circle with a diameter d = 1 (and a radius r = 1.2), enclosed by a square with side length 1. We know that the area A of the circle is:
$$A = r^{2}*\pi = \frac{1}{2}*\frac{1}{2}*\pi = \frac{\pi}{4} $$


We now sample random coordinates from the normal distribution N(0, 1).
The points can vary from (0, 0) to (1, 1). A number of points will fall inside
of the circle, whilst other points will fall outside of the circle. We keep track
of the number of points inside of the circle N_inner, and of the total number
of points N_total.


If we divide the number of points that fall inside of the circle by the total
number of points, we get an approximation of the circle’s area:

$$ \frac{N_{inner}}{N_{total}}≈ \frac{Area  of  Circle}{Area  of  Square}= \frac{A}{1} = A $$


Therefore, if we combine equation (1) and (2), we obtain an approximation of π:

$$ \frac{\pi}{4} ≈ \frac{N_{inner}}{N_{total}}⇔ \pi ≈ \frac{N_{inner}}{N_{total}}*4 $$


For example, in Figure 1 below we have 786 sampled points inside of the circle(red) out of 1000 sampled points (black + red) which results in:


$$ \pi ≈ \frac{N_{inner}}{N_{total}}* 4 = \frac{786}{1000}* 4 = 3.144 $$

![Failed to load the picture](https://www.101computing.net/wp/wp-content/uploads/estimating-pi-monte-carlo-method.png)
Figure 1: A total of 1000 sampled points (black + red). 786 fell inside of the
circle (red), while the other 214 fell outside of the circle (black).

This example shows the power of sampling a large number of random
values within a certain domain, and using the result to make some sort of
approximation. If we extend the number of generated points in the example
above, to let’s say 1 000 000, then the result would be significantly more
accurate. The constant $\pi$ can be calculated “infinitely” precise by the perfect circle’s circumference and diameter. Other more complex problems on
the other hand, such as stochastic processes, can’t be calculated just as an
expression of numbers. In those problems the power of randomness can be
truly useful in order to obtain a good enough approximation of the problem
at hand.

```python
def get_company_info (ticket, starting_period, ending_period = None, source = 'yahoo'):
    return data.DataReader(ticket, source, starting_period, ending_period)
```

```python
company_data = get_company_info('AAPL','1/1/1990','18/5/2021')
```

Now, let's calculate the volatility and compound annual growth rate(CAGR).


determining the annual volatility of returns. 
As most people in finance understand it, volatility is standard deviation of returns

What CAGR can tell you:


The compound annual growth rate isn't a true return rate, but rather a representational figure. It is essentially a number that describes the rate at which an investment would have grown if it had grown the same rate every year and the profits were reinvested at the end of each year. In reality, this sort of performance is unlikely. However, CAGR can be used to smooth returns so that they may be more easily understood when compared to alternative investments.

To calculate the CAGR of an investment:
<ol>
    <li>Divide the value of an investment at the end of the period by its value at the beginning of that period.</li>
    <li>Raise the result to an exponent of one divided by the number of years.</li>
    <li>Subtract one from the subsequent result.</li>
    </ol>

```python
def calc_statistical_variables(company_data, number_of_trading_days = 252):
    
    # cagr calculations
    time_elapsed = (company_data.index[-1] - company_data.index[0]).days
    
    price_ratio = (company_data['Adj Close'][-1] / company_data['Adj Close'][1])
    inverse_number_of_years = 1 / (time_elapsed / 365.0)
    cagr = price_ratio ** inverse_number_of_years - 1
    
    #determining the annual volatility of returns. 
    #As most people in finance understand it, volatility is standard deviation of returns.
    vol = company_data['Adj Close'].pct_change().std()
    
    #In order to make the volatilty useful as an annual measure, it has to be scaled by an annualization factor.
    #This factor is most often the square root of the number of trading days in a year which happens to be 252.
    vol = vol * math.sqrt(number_of_trading_days)
    
    return vol,cagr
```

```python
vol, cagr = calc_statistical_variables(company_data)
```

```python
print ("cagr (mean returns) : ", str(round(cagr,4)))
print ("vol (standard deviation of return ) :", str(round(vol,4)))
```

### 2. Generate Random Values


Perfect! This is all we need to model a standard deviation of stock prices. Now, let's actually create a random sample using these parameters. For this, we'll use the normal function from the numpy library.

```python
def make_random_walk(company_data, cagr, vol, number_of_trading_days=252):
# generating random values

#np.random.normal(first : Mean ("centre") of the distribution,
#second: Standard deviation (spread or "width") of the distribution. Must be non-negative.
#third size :Output shape.)
# said formal, it generates number_of_trading_days times random "coefficients" of the mean, using the standard deviation as parameter
    daily_return_percentages = np.random.normal((1+cagr)**(1/number_of_trading_days), vol/math.sqrt(number_of_trading_days),number_of_trading_days)
    
    # initializing price_series with the last price value of the real data
    price_series = [company_data['Adj Close'][-1]]

    # int the loop we 
    for drp in daily_return_percentages:
        # the result "coefficient" is multiplied by the last value of the price_seies and then appended to price_series,
        # and in the next iteration the last appended value will be used as multiplier with the coefficient
        price_series.append(price_series[-1] * drp)
    
    plt.plot(price_series)
    plt.title("Random walk of the price")
    plt.xlabel("days")
    plt.ylabel("price")
    plt.show()
    return price_series
```

```python
random_walk = make_random_walk(company_data, cagr, vol, 252)
```

```python
plt.hist(random_walk,bins=50)
plt.title("Price distribution")
plt.xlabel("Closing price")
plt.ylabel("Count")
plt.show()
```

### 3. Apply them over many, many trials

```python
trials_of_random_walk_distribution = make_many_random_walks(company_data, cagr, vol)
```

### 4. Collect Results and Analyze

```python
expected_ending_price_point = round(np.mean(trials_of_random_walk_distribution),2)
print("Expected Ending Price Point : ", str(expected_ending_price_point))
```

```python
population_mean = (cagr+1) * company_data['Adj Close'][-1]
print ("Sample Mean : ", str(expected_ending_price_point))
print ("Population Mean: ", str(round(population_mean,2)));
print ("Percent Difference : ", str(round((population_mean - expected_ending_price_point)/population_mean * 100,2)), "%")
```

```python
pct_5_quantile = np.percentile(trials_of_random_walk_distribution,5) # calculating the first 5% of the left tail of the distribution
pct_95_quantile = np.percentile(trials_of_random_walk_distribution,95) # calculating the first 5% of the right tail of the distribution
mean = np.percentile(trials_of_random_walk_distribution, 50); # calculating the mean of the distribution
print("5% quantile =", pct_5_quantile)
print("95% quantile =", pct_95_quantile)
print ("Mean of sample: ", str(round(mean,2)))
```

## 8. Monte Carlo method compared to Autoregressive Integrated Moving Average model (ARIMA)

<a id="8"></a>

The methods use different input data in predicting stock prices, including different time horizons - ARIMA model uses time series data to predict future prices, while Monte Carlo simulation uses random walk component.

There are papers which says that there is a slight advantage in accuracy of predictions of ARIMA model for shorter periods  (5 days), while MCs has undoubted advantage for longer periods (more than 21days).

It is known that in shorter periods stock prices have elements of predictability, what questions the level of efficiency of the world leading stock exchange. In longer periods (21 days) results indicate market efficiency, i.e. model with random walk component (MCs) has advantage compared to the model based on historical prices (ARIMA).
